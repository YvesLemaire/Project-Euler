On cherche le résultat $r$ sous la forme de sa décomposition en facteurs premiers.

$$r=\prod_{p\; \text{ premier}}p^{\alpha_p} $$ $$\text{où } \alpha_p=\max_{p^{\alpha}\leqslant n}\alpha$$.

```python
from projectEuler import eratosthene

def f(n):
    r = 1
    primes = eratosthene(n)
    for p in primes:
        q = p
        while q <= n:
            q *= p
            r *= p
    return r

f(20)
```
