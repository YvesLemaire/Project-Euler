Notons $\sigma_p(n)$ la somme des puissances $p^{\text e}$ des chiffres de $n$.

On cherche la liste $f(p)$ des entiers $n$ qui vérifient $\sigma_p(n)=n$. 

Si $n$ comporte $k$ chiffres alors $\sigma_p(n)\leqslant k9^p$ et $10^{k-1}\leqslant n$.  
Donc, si l'on veut que $\sigma_p(n)=n$, il faut que $k\leqslant K=$ la valeur maximale de $k$ qui vérifie $k9^p\geqslant 10^{k-1}$ et il faut
aussi que $n=\sigma_p(n)\leqslant k9^p\leqslant M=K9^p$

```python
def f(p):
    K = 0
    while (K + 1) * 9 ** p >= 10 ** K:
        K += 1
    M = K * 9 ** p
    L = []
    dict = {c : int(c) ** p for c in '0123456789'}
    for n in range(2, M):
        if sum(map(lambda c: dict[c], str(n))) == n:
            L.append(n)
    return L

sum(f(5)) 
```
