```python
from functools import cache 

f = lambda n: 3 * n + 1 if n % 2 else n // 2

@cache # memoize la fonction suivante
def chain_length(n):
    return 1 if n == 1 else 1 + chain_length(f(n))

n0 = 1
m = 1
for n in range(1, 1000001):
    l = chain_length(n)
    if l > m:
        m = l
        n0 = n

n0
```
