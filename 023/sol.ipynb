```python
from projectEuler import sumDivisors

N = 28123
abundance = lambda n: sumDivisors(n) - 2 * n
abundants = [n for n in range(1, N) if abundance(n) > 0]
halfAbundants = [n for n in abundants if n <= (N - 1) // 2]

t = [False] * N # t[n] = True <=> n est somme de 2 abondants
for n in halfAbundants:
    for m in abundants:
        if n <= m < N - n:
            t[n + m] = True
sum(s for s in range(1, N) if not t[s])
```

```python
import numpy as np
from projectEuler import sumDivisors

N = 28123
abundance = lambda n: sumDivisors(n) - 2 * n
abundants = [n for n in range(1, N) if abundance(n) > 0]

A = np.array(abundants)
B = np.array(range(1, N + 1))
adds = np.add.outer(A, A)
C = np.unique(adds[adds <= N])
int(np.sum(B) - np.sum(C))
```
