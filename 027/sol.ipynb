```python
from projectEuler import eratosthene

M = 13000
primes = eratosthene(M)
primes_bis = eratosthene(1000)

# 12s
N = 0
for b in primes_bis:
    for a in range(-999, 1000):
        n = 0
        v = b # v = n^2 + a n + b
        while v in primes:
            n += 1
            v += a + 2 * n - 1
            if v > M: raise Exception("Il faut augmenter M")
        if n > N:
            N = n
            A = a
            B = b
N, A , B, A * B
```
