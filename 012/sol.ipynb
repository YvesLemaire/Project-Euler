Le nombre de diviseurs d'un entier $m=p_1^{\alpha_1}p_2^{\alpha_2}\ldots p_k^{\alpha_k}$ ($p_i$ premiers) est $d(m)=(\alpha_1+1)(\alpha_2+1)\ldots(\alpha_k+1)$.

Le $n^{\text e}$ nombre triangulaire est $\displaystyle t_n=\frac{n(n+1)}{2}=\overline{n}\times\overline{n+1}$ oÃ¹ $\overline{n}=n$ si $n$ est impair et $n/2$ si $n$ est pair.

Comme $n$ et $n+1$ sont premiers entre eux, $\overline{n}$ et $\overline{n+1}$ le sont aussi ; donc $d(t_n)=d(\overline{n})\times d\left(\overline{n+1}\right)$.

```python
from projectEuler import primeDecomposition

n = 0
d = 0 # d(bar(n))
nEven = True
while True:
    n += 1
    nEven = not nEven
    c = d 
    f = primeDecomposition(n + 1 if nEven else (n + 1) // 2)
    v = [alpha for _, alpha in f] # the alpha_i's for the number n + 1 or (n+1)/2
    d = 1
    for alpha in v:
        d *= alpha + 1
    if c * d >= 500:
        break
n * (n + 1) // 2
```
