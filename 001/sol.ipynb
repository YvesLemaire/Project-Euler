```python
# solution en O(n) (n = 1000)

sum(range(3, 1000, 3)) + sum(range(5, 1000, 5)) - sum(range(15, 1000, 15))
```

```python
# solution en O(1)

def arithmeticSum(p):
    # -> la somme des multiples de p < a 1000
    n = 999 // p
    return p * n * (n + 1) // 2

arithmeticSum(3) + arithmeticSum(5) - arithmeticSum(15)
```
