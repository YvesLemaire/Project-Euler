On numérote les jours en considérant le $1/1/1900$ comme le jour $0$.

```python
def is_leap(Y):
    return Y % 4 == 0 and (Y % 100 != 0 or Y % 400 == 0)

def leap(q, Y): # nombre de multiples de q entre 1900 et Y - 1
    return (Y - 1)// q - 1899 // q

def first(Y): # numero du 1/1/Y
    return 365 * (Y - 1900) + leap(4, Y) - leap(100, Y) + leap(400, Y)

days = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

def num_of_date(d, m, y):
    return first(y) + sum(days[:m - 1]) + d - (0 if is_leap(y) and m > 1 else 1)

c = 0
for y in range(1901, 2001):
    for m in range(12):
        if num_of_date(1, m, y) % 7 == 6:
            c += 1
c
```
