```python
from projectEuler import isPalindrome

r = 0
for p in range(100, 1000):
    for q in range(p, 1000):
        n = p * q
        if n > r and isPalindrome(n):
            r = n
r

```

Le résultat est de la forme   
$r=abccba=100000\times a + 10000\times b + 1000\times c + 100\times c + 10\times b + a$  
$=11(9091\times a + 910\times b + 100\times c)$ est divisible par $11$ ; donc $p$ ou $q$ aussi.
